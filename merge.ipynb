### **1. "Brincando" com o método para conhece-lo.**

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'ticker': ['petr4', 'vale3', 'itub4'], 'setor': ['óleo e gás', 'mineração', 'financeiro']})
df1
```

```python
df2 = pd.DataFrame({'ticker': ['vale3', 'itub4', 'goll4'], 'tendencia': ['alta', 'baixa', 'consolidação']})
df2
```

```python
pd.merge(df1, df2, how='inner', on='ticker')
```

```python
df1 = pd.DataFrame({'ticker': ['petr4', 'vale3', 'itub4'], 'setor': ['óleo e gás', 'mineração', 'financeiro']})
df2 = pd.DataFrame({'tickers': ['vale3', 'itub4', 'goll4'], 'tendencia': ['alta', 'baixa', 'consolidação']})
pd.merge(df1, df2, how='inner', left_on='ticker', right_on='tickers')
```

```python
pd.merge(df1, df2, how='inner', left_index=True, right_index=True)
```

### **2. Exemplo de mercado financeiro**

```python
data = pd.date_range(start='2023-07-01', periods=5, freq='D')
cotacoes = [100000, 100150, 99950, 100300, 100100]

indice = pd.DataFrame(cotacoes, index=data, columns=['indice futuro'])
indice
```

```python
data = pd.date_range(start='2023-07-03', periods=5, freq='D')
cotacoes = [5.15, 5.21, 5.18, 5.35, 5.30]

dolar = pd.DataFrame(cotacoes, index=data, columns=['dolar futuro'])
dolar
```

```python
pd.merge(indice, dolar, how='inner', left_index=True, right_index=True)
```
