# **Rastreador de Ações com Python**

## **1. Instalando e importando bibliotecas**

Usaremos a **Yfinance** para obter os dados dos ativos financeiros e o **matplotlib** para gerar os gráficos.

```python
!pip install yfinance --quiet
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
```

## **2. Definir as ações da carteira**

Aqui escolhemos as ações que desejamos rastrear e definimos seus símbolos como elementos de uma lista. **Se for ação brasileira, adicione ".SA" ao final.** Lembrando que **cada ação terá o valor da moeda local.**

```python
acoes = ['AAPL', 'MSFT', 'PETR4.SA', 'BBDC4.SA']
```

## **3. Importar dados dos ativos financeiros**

Agora, vamos usar a biblioteca Yfinance para obter os dados das ações. Adicione o seguinte código:

```python
dados_acoes = yf.download(acoes, period='1d', start='2022-01-01')
```

```python
dados_acoes
```

## **4. Criando o primeiro gráfico**

Vamos plotar os **preços de fechamento** das ações no gráfico.

```python
dados_acoes['Adj Close'].plot()
plt.xlabel('Data')
plt.ylabel('Preço')
plt.title('Rastreador de Ações')
plt.legend(fontsize=7)
plt.show()
```

## **5. Normalizando os preços**

Como cada ativo parte de um valor inicial diferente, vamos normalizar para que **todos partam do mesmo ponto**, a fim de obter uma comparação mais precisa.

Esse código normaliza os preços de fechamento das ações **dividindo-os pelo preço de fechamento inicial de cada ação.** Em seguida, cria um novo gráfico com os preços normalizados.

```python
dados_normalizados = dados_acoes['Adj Close'] / dados_acoes['Adj Close'].iloc[0]
dados_normalizados.plot()
plt.xlabel('Data')
plt.ylabel('Preço')
plt.title('Rastreador de Ações - Normalizado')
plt.legend(fontsize=7)
plt.show()
```
